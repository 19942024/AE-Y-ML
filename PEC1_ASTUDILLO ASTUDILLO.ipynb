2. Identificar, definir y explicar para este caso los principales elementos de un contraste de hiptesis.

```r
categ_gan_perd = function(transacciones) {
  # Usamos ifelse para asignar una categora basada en si el valor es positivo, negativo o cero
  categorias = ifelse(transacciones > 0, "Ganancia",
                  ifelse(transacciones < 0, "Prdida", "Neutral"))
  return(categorias)
}
```

```r
# Probamos la funcin para el vector de ejemplo
x = categ_gan_perd(c(100, -50, 0, 25))
x
```

Usamos function para crear una funcin propia. Definimos el parmetro de entrada, que en este caso ser transacciones, el vector numrico de ganancias y prdidas o neutral.
luego usamos la funcin ifelse que permite evaluar una condicin para cada elemento de un vector y devolver un resultado correspondiente para cada elemento.

```r
Bono_pagado = function(aos_servicio, salario_actual) {
  # Calcular el bono como 10% del salario por cada ao de servicio
  Bono_final = (aos_servicio) * (0.10 * salario_actual)
  
  # definimos el lmite mximo del bono
  bono_maximo = salario_actual * 0.50
  
  # Si el Bono final  es mayor el bono mximo,igualar al bono mximo
  if (Bono_final > bono_maximo) {
    Bono_final= bono_maximo
  }
  
  return(Bono_final)
}

# Ejemplo de uso de la funcin
Bono_pagado(5, 1000)


```

Usamos function para crear la funcin propia . Definimos dos parmetro de entrada, que en este caso ser aos de servicio y salario actual.
luego usamos la funcin if para verificar si el bono calculado excede el lmite mximo. Si es as, ajustamos el bono al lmite mximo.
