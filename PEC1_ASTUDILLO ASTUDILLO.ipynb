### 02 Ejercicio [1.25 puntos] 

Considerar dos conjunto de datos que se ajustan a las siguientes distribuciones normales, $X_1∼N(μ_1,σ_1)$ y $X_2∼N(μ_2,σ_2)$. Supongamos que $x_1=(10, 12, 15, 14, 11, 13, 12)$ y $x_2=(14, 15, 16, 17, 13, 14, 16, 15)$ son muestras de los tiempos de respuesta (en días) a dos tratamientos médicos diferentes, $X_1$ y $X_2$, respectivamente. Se solicita:


1. Calcular el p-valor del contraste $H_0:μ_1=μ_2$ contra $H_1:μ_1<μ_2$ y justificar si podemos rechazar la hipótesis nula en favor de la alternativa para un nivel de significación de $\alpha=0,05$. 


2. Identificar, definir y explicar para este caso los principales elementos de un contraste de hipótesis. 


3. Determinar si el contraste es paramétrico o no paramétrico y justificar la respuesta. Considerar los análisis previos oportunos para cada caso (normalidad, varianza, etc.) para seleccionar el test adecuado. 

<div style="background-color:#F7DC6F"><b>Resolución Ejercicio 02</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

```r
data_x1=c(10,12,15,14,11,13,12)
data_x2=c(14,15,16,17,13,14,16,15)
data_x1
data_x2
# mostramos los datos de las muestras 
  
```

```r
 t.test(data_x1,data_x2, alternative = "less", var.equal = FALSE) 
# el p-valor es 0.003973 el cual es menor al nivel de significancia, 
#por lo tanto rechazamos la hipotesis nula.
```

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

La hipotesis nula asegura que no existe diferencia en media de tiempo de respuesta a los medicamentos, mientras la hipotesis alternativa afirma que la media del tratamiento 1 es menor al tratamiento dos. al realizar la prueba t.test rechazamos la hipotesis nula y comprobamos que la media de tiempo del tratamiento 1 es menor al tratamiento 2.

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

```r
bartlett.test(list(data_x1,data_x2)) #mediante este test comprobamos que es un contraste parametrico 
```

Sabemos que tenemos dos variables diferentes con una distribucion normal, entonces para confirmar si es un contraste parametrico o no , realizamos la prueba bartlett.test la cual nos muestra si tienen homogeneidad de varianza. En este caso nos dio un p-value de 0.505 lo cual nos confirma que es un contraste parametrico y usamos la prueba test t.

```r
categ_gan_perd = function(transacciones) {
  # Usamos ifelse para asignar una categoría basada en si el valor es positivo, negativo o cero
  categorias = ifelse(transacciones > 0, "Ganancia",
                  ifelse(transacciones < 0, "Pérdida", "Neutral"))
  return(categorias)
}
```

```r
# Probamos la función para el vector de ejemplo
x = categ_gan_perd(c(100, -50, 0, 25))
x
```

Usamos function para crear una función propia. Definimos el parámetro de entrada, que en este caso será transacciones, el vector numérico de ganancias y pérdidas o neutral.
luego usamos la función ifelse que permite evaluar una condición para cada elemento de un vector y devolver un resultado correspondiente para cada elemento.

```r
Bono_pagado = function(años_servicio, salario_actual) {
  # Calcular el bono como 10% del salario por cada año de servicio
  Bono_final = (años_servicio) * (0.10 * salario_actual)
  
  # definimos el límite máximo del bono
  bono_maximo = salario_actual * 0.50
  
  # Si el Bono final  es mayor el bono máximo,igualar al bono máximo
  if (Bono_final > bono_maximo) {
    Bono_final= bono_maximo
  }
  
  return(Bono_final)
}

# Ejemplo de uso de la función
Bono_pagado(5, 1000)


```

Usamos function para crear la función propia . Definimos dos parámetro de entrada, que en este caso será años de servicio y salario actual.
luego usamos la función if para verificar si el bono calculado excede el límite máximo. Si es así, ajustamos el bono al límite máximo.
